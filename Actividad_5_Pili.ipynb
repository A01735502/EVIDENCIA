```python
pat_cdmx = r'(ciudad\s+de\s+m[ée]xico|cdmx|mexico\s+city|m[ée]xico\s*,?\s*d\.?\s*f\.?)'

# Colapsa TODAS las filas que contengan alguna variante a una sola etiqueta
mask = df['host_location'].str.contains(pat_cdmx, case=False, na=False, regex=True)
df.loc[mask, 'host_location'] = 'CDMX'

# Opcional: normaliza Unknown
df['host_location'] = df['host_location'].fillna('Unknown')
df['host_location'] = df['host_location'].replace(r'^\s*(nan|none|unknown)?\s*$', 'Unknown', regex=True)

# Recuenta y grafica
filtro = df['host_location'].value_counts().reset_index()
filtro.columns = ['host_location', 'count']
filtro = filtro[filtro['count'] > 200]

ax = filtro.plot(kind='bar', x='host_location', y='count', legend=False, color=['slateblue', 'teal', 'mediumorchid', 'darkmagenta', 'midnightblue'])
ax.set_title('Host Location')
ax.set_ylabel('Frecuencia')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
```
